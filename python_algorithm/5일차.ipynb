```python
# 최적의 풀이
import random
import itertools

# 컴퓨터가 맞추기 위해서 어떤 방법을 쓰면 좋을까요?
#  1. 사람이라면 어떤 식으로 숫자를 맞출까?
#    1) 그냥 랜덤하게 뿌릴까요?
#    2) 답 상관 없이 랜덤하게 뿌리는 것은 안 되니..
#  2. 숫자를 생각해서 족보에 만족하면 해당 숫자를 보여줍니다.

# strike와 ball을 계산하는 함수
def judge(a, b):
    strike, ball = 0, 0
    for i in range(3):
        if a[i] == b[i]: strike += 1
        if a[i] in b: ball += 1
    return strike, ball-strike

# 추측을 할 수 있는 모든 경우를 guessList에 저장
t = [ i for i in range(1, 10) ]

guessList = [ k for k in itertools.permutations(t, 3) ]

# 랜덤하게 추측하도록 미리 guessList를 섞는다
random.shuffle(guessList)
round = 1
history = [ ]

for guess in guessList:
    isOk = True
    for h in history:
        s, b = judge(h[0], guess)
        
        if s != h[1] or b != h[2]:
            isOk = False
            break
            
    if not isOk: continue
        
    # 추측한 숫자를 출력
    print("Guess :", "".join(map(str, guess)))
    
    s, b = map(int, input().split())
    
    if s == 3:
        print("Score : %d"%round)
        break
        
    history += [ (guess, s, b) ]
    rount += 1
```

```python
# 하노이탑
step = 1

def dec2bin(x):
    s = ""
    while x > 0:
        s = str(x%2) + s
        x //= 2
    return s

def hanoi(n, a, b, c):
    global step
    if n == 1:
        x = a[1].pop()
        b[1].append(x)
        print("%s: %s --> %s (%d)"%(dec2bin(step), a[0], b[0], x))
        step += 1
        return
    hanoi(n-1, a, c, b)
    hanoi(1, a, b, c)
    hanoi(n-1, c, b, a)

n = int(input("하노이 타워의 원반 수는 : "))
stackA = [ n-i for i in range(n) ]
hanoi(n, ("A", stackA), ("B", []), ("C", []))
```
